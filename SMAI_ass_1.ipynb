```python
from random import choice
from numpy import array, dot, random
```

```python
unit_step = lambda x: 0 if x < 0 else 1
```

```python
# w = random.rand(9)
w = np.zeros(9)
errors = [] 
eta = 0.2
n = 5*len(data)
```

```python
w
```

```python
#Vanila perceptron
errors = []
for i in xrange(n):
    random_sample = choice(data) 
    x, expected =random_sample[:-1],random_sample[-1]
    result = dot(w, x)
    error = expected - unit_step(result) 
    errors.append(error)
    w += eta * error * x
```

```python
#Voted Perceptron
errors = []

votes = [[w,0]]
count=0
#pairs of w,count where count represents 
#how many times current weight w has survived

for i in xrange(n):
    random_sample = choice(data) 
    x, expected =random_sample[:-1],random_sample[-1]
    result = dot(w, x)
    error = expected - unit_step(result) 
    if error==0:
        votes.append([w,count])
        count=1
    else:
        count=count+1
    errors.append(error)
    w += eta * error * x
```

```python
#inferencing
Y_pred = []
for sammple in data: 
    x = sammple[:-1]
    w_star = np.array(votes)[np.argmax(vt[:,1])][0]
    result = dot(x, w_star)
    Y_pred.append(unit_step(result))
    print("{}: {} -> {}".format(sammple[-1], result, unit_step(result)))
```

```python
#Compute Accuracy

Y = data[:,-1]
acc=0
for i,j in zip(Y,Y_pred):
    if i==j:
        acc+=1
acc = float(acc)/len(Y)
print acc
```
